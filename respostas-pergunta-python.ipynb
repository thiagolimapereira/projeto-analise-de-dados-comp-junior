```python
'''
A abordagem que utilizei nesta questão foi, primeiramente, verificar a existência de tuplas duplicadas, as quais se mostraram inexistentes.
Em seguida, verifiquei a presença de valores nulos no dataset e decidi analisar em que contexto esses valores nulos estavam inseridos.

Ao meu ver, existiam duas situações em que valores nulos apareciam. A primeira ocorria quando o status do pedido era "cancelado", o que
considero normal, uma vez que os pedidos podem ser cancelados a qualquer momento, inclusive antes de serem finalizados. Já a segunda situação
ocorria quando o status do pedido não era "cancelado", o que sugere possíveis problemas na atualização do banco de dados (o que torna os dados
pouco confiáveis) ou problemas na entrega, algo que não é registrado em nosso banco de dados, tornando esses registros inapropriados para análise.

Portanto, decidi excluir as tuplas que possuem valores nulos, mas que não têm o status de "cancelado" no dataset.
'''


#verificando a existência de valores duplicados no dataset
print(df.duplicated().sum())
print('\n------------------------------------\n')

#verificando e identificando a existência de valores nulos no dataset
print(df.isnull().sum())
print('\n------------------------------------\n')

#analisando as tuplas com valores nulos no dataset
#df[df.isnull().any(axis=1)].head(20)

#identificando os valores nulos no dataset para tuplas que não possuem order_status como 'canceled' para melhor análise
#df.loc[df['order_status'] != 'canceled'].isnull().sum()

#excluindo as tuplas com valores nulos, mas que não possuem order_status como 'canceled'
df = df[(df['order_status'] == 'canceled') | df.notna().all(axis=1)]

#identificando os valores nulos no dataset após a exclusão das tuplas
print(df.isnull().sum())
```

## Exercício 05
Visualizações claras podem revelar tendências que antes não eram possíveis de serem capturadas. Crie um gráfico de barras vertical que mostre o volume de vendas mensal por estado. Use essa visualização para identificar quais estados apresentaram maior crescimento no último ano.

```python
#criando as colunas 'year' e 'month' para armazenar o ano e o mês da compra
df['year'] = df['order_purchase_timestamp'].dt.year
df['month'] = df['order_purchase_timestamp'].dt.month

#agrupando os dados pelo estado, ano e mês e contando o número de vendas
pivot_table = df.groupby(['customer_state', 'year', 'month']).size().reset_index(name='monthly_sales')

#filtrando os dados para considerar apenas os meses consistentes de 2018 (setembro teve apenas uma venda registrada)
pivot_table = pivot_table[(pivot_table['year'] == 2018) & (pivot_table['month'] <= 8)]

#criando a tabela pivô com os meses como colunas e preenchendo os valores ausentes com 0
pivot_table = pivot_table.pivot_table(index='customer_state', columns='month', values='monthly_sales', fill_value=0)

#ordenando os estados em ordem decrescente de quantidade de vendas
pivot_table['quantidadeVendas'] = pivot_table.sum(axis=1)
pivot_table = pivot_table.sort_values(by='quantidadeVendas', ascending=False)
pivot_table.drop('quantidadeVendas', axis=1, inplace=True)

#criando o gráfico de barras verticais empilhadas
pivot_table.plot(kind='bar', stacked=True, figsize=(12, 8))

#formatando o gráfico
plt.xlabel('Estado')
plt.ylabel('Volume de Vendas')
plt.title('Volume de Vendas Mensal por Estado (2018)')
plt.legend(title='Mês', bbox_to_anchor=(1.05, 1), loc='upper left')

#exibindo o gráfico
plt.show()

'''
Pelo gráfico de barras verticais solicitado, não é tão simples identificar quais estados apresentaram crescimento no
volume de vendas mensais no último ano, apenas quais possuem maiores volumes de vendas, já que não é fácil comparar a
diferença no tamanho das divisões das barras. Consigo perceber um leve aumento no volume de vendas em São Paulo, mas
não consigo identificar com clareza se houve um aumento significativo em outros estados. Acredito que o ideal seria
utilizar um gráfico de linhas para comparar o volume de vendas mensais ao longo do tempo para cada estado, o que
facilitaria a identificação de tendências e variações. Então, creio que um gráfico de barras verticais não seja uma
visualização clara para o propósito do exercício.
'''
```

## Exercício 06
Para incentivar uma competição saudável entre os vendedores, você decide analisar o desempenho individual. Calcule e visualize o total de vendas por vendedor usando um gráfico de barras horizontal no Matplotlib, destacando o vendedor com maior volume de vendas.

```python
#agrupando os dados pelo vendedor e somando o total em vendas
total_vendas_por_vendedor = df.groupby('seller_id')['price'].sum().reset_index()

#agrupando os dados pelo vendedor e contando o volume de vendas
quantidade_vendas_por_vendedor = df['seller_id'].value_counts().reset_index()

#fazendo o merge dos dois dataframes
total_vendas_por_vendedor = pd.merge(total_vendas_por_vendedor, quantidade_vendas_por_vendedor, on='seller_id')

#ordenando os dados pelo total em vendas
total_vendas_por_vendedor = total_vendas_por_vendedor.sort_values(by='price', ascending=False)

#encontrando o vendedor com o maior volume de vendas
vendedor_mais_vendas = total_vendas_por_vendedor.loc[total_vendas_por_vendedor['count'].idxmax()]

#selecionando os 20 vendedores com maiores totais em vendas
topVendedores = total_vendas_por_vendedor.head(20).sort_values(by='price')

#criando o gráfico de barras horizontais
plt.figure(figsize=(12, 6))
plt.barh(topVendedores['seller_id'], topVendedores['price'], color='skyblue')

#destacando o vendedor com o maior volume de vendas
plt.barh(vendedor_mais_vendas['seller_id'], vendedor_mais_vendas['price'], color='orange')
plt.text(vendedor_mais_vendas['price'], vendedor_mais_vendas['seller_id'], f'Vendedor com Maior Volume de Vendas ({vendedor_mais_vendas["count"]} Vendas) ', color='black', fontsize=8,va='center', ha='right')

#formatando o gráfico
plt.xlabel('Total em Vendas($)')
plt.ylabel('ID do Vendedor')
plt.title('Total em Vendas($) por Vendedor')

#exibindo o gráfico
plt.show()
```

## Exercício 08
A precisão nas previsões de entrega é um indicador chave de desempenho logístico. Desenvolva um método para comparar as datas estimadas e reais de entrega, identificando padrões de atrasos ou adiantamentos nas entregas. Faça um gráfico de linhas com o tempo médio de atraso por semana ao longo do tempo. O eixo y deve ser o tempo médio de atraso (em dias) e o eixo x as semanas do ano. 

```python
#fazendo uma coluna para armazenar a diferença entre a data de entrega estimada e a data de entrega real (negativo se a entrega foi adiantada e positivo se a entrega foi atrasada)
df['diferencaEntreEsperadoEReal'] = (df['order_delivered_customer_date'] - df['order_estimated_delivery_date']).dt.days.astype('Int64')

#verificando a média da diferença entre a data de entrega estimada e a data de entrega real
print('Média da diferença entre a data de entrega estimada e a data de entrega real: ', round(df['diferencaEntreEsperadoEReal'].mean(),2), '\n(um valor negativo indica que a entrega foi adiantada em x dias)')

print('\n------------------------------------\n')

#verificando, para cada estado, a média da diferença entre a data de entrega estimada e a data de entrega real e ordenado de forma que os que mais adiantam (ou menos atrasm) apareçam no topo
print('Média da diferença entre a data de entrega estimada e a data de entrega real por estado:\n(Valores negativos indicam que a entrega foi adiantada)')
print(round(df.groupby('customer_state')['diferencaEntreEsperadoEReal'].mean(),2).sort_values())
print('\n------------------------------------\n')

#verificando, para pedidos que atrasam, quanto tempo, em média, eles atrasam
print('Pedidos que atrasam, atrasam, em média,', str(round(df[df['diferencaEntreEsperadoEReal'] > 0]['diferencaEntreEsperadoEReal'].mean(), 2)), 'dias')

#verificando, para pedidos que atrasam, qual a média do preço do produto e preço do frete
print('Pedidos que atrasam, possuem um preço médio de produto de R$', str(round(df[df['diferencaEntreEsperadoEReal'] > 0]['price'].mean(), 2)), 
      ' e um preço médio de frete de R$', str(round(df[df['diferencaEntreEsperadoEReal'] > 0]['freight_value'].mean(), 2)))

#verificando a porcentagem de pedidos que atrasam por estado
print('Porcentagem de pedidos que atrasam por estado:')
print((round((df[df['diferencaEntreEsperadoEReal'] > 0].groupby('customer_state').size() / df.groupby('customer_state').size())*100, 2)).sort_values(ascending=False).astype(str) + '%')

print('\n------------------------------------\n')


#verificando, para pedidos que adiantam, quanto tempo, em média, eles adiantam
print('Pedidos que adiantam, adiantam, em média,', str(round(df[df['diferencaEntreEsperadoEReal'] < 0]['diferencaEntreEsperadoEReal'].mean()*-1, 2)), 'dias')

#verificando, para pedidos que adiantam, qual a média do preço do produto e preço do frete
print('Pedidos que adiantam, possuem um preço médio de produto de R$', str(round(df[df['diferencaEntreEsperadoEReal'] < 0]['price'].mean(), 2)), 
      ' e um preço médio de frete de R$', str(round(df[df['diferencaEntreEsperadoEReal'] < 0]['freight_value'].mean(), 2)))

#verificando a porcentagem de pedidos que adiantaram por estado
print('Porcentagem de pedidos que adiantam a entrega por estado:')
print((round((df[df['diferencaEntreEsperadoEReal'] < 0].groupby('customer_state').size() / df.groupby('customer_state').size())*100, 2)).sort_values(ascending=False).astype(str) + '%')

#fazendo uma correlação de pearson entre a diferença entre a data de entrega estimada e a data de entrega real e o preço do produto
correlacao = df[['diferencaEntreEsperadoEReal', 'price']].corr(method='pearson')
plt.figure( figsize=(7, 3) )
sns.heatmap( correlacao, annot=True, cbar=False )

#fazendo uma correlação de pearson entre a diferença entre a data de entrega estimada e a data de entrega real e o preço do frete
correlacao = df[['diferencaEntreEsperadoEReal', 'freight_value']].corr(method='pearson')
plt.figure( figsize=(7, 3) )
sns.heatmap( correlacao, annot=True, cbar=False )


#criando uma nova coluna para armazenar a semana do ano em que o pedido foi feito
df['week_of_year'] = df['order_purchase_timestamp'].dt.isocalendar().week

#calculando a média do tempo de atraso por semana ao longo do tempo
mediaAtrasoPorSemana = df[df['diferencaEntreEsperadoEReal'] > 0 ].groupby('week_of_year')['diferencaEntreEsperadoEReal'].mean()

#criando o gráfico de linhas
plt.figure(figsize=(12, 6))
plt.plot(mediaAtrasoPorSemana.index, mediaAtrasoPorSemana.values, marker='o')

#formatando o gráfico
plt.xlabel('Semana do Ano')
plt.ylabel('Tempo Médio de Atraso (dias)')
plt.title('Tempo Médio de Atraso por Semana ao Longo do Tempo')
plt.grid(True)

#exibindo o gráfico
plt.show()


'''
Com base nos dados analisados, podemos observar que as empresas entregam, em média, 12 dias antes da data estimada de entrega.

Os estados que as empresas mais adiantam a entrega, em média, em relação a estimativa de entrega são, respectivamente: Acre,
Roraima e Amazonas.

Os pedidos que atrasam, atrasam, em média, 10,49 dias.
Os pedidos que atrasam possuem um preço médio de produto de R$135,71 e um preço médio de frete de R$22,71.
A porcentagem de pedidos que atrasam são as maiores, respectivamente, nos estados Alagoas, Maceió e Sergipe.

Os pedidos que adiantam, adiantam, em média, 13,81 dias.
Os pedidos que adiantam possuem um preço médio de produto de R$118,86 e um preço médio de frete de R$19,76.
A porcentagem de pedidos que adiantam a entrega são as maiores, respectivamente, nos estados Acre, Amazonas e Roraima.

Parece não haver uma correlação entre a diferença entre a data de entrega estimada e a data de entrega real e o preço do produto,
bem como entre a diferença entre a data de entrega estimada e a data de entrega real e o preço do frete.

A semana do ano em que os pedidos mais atrasam é a semana 22 (final de Maio e começo de Junho), enquanto a semana do ano em que
os pedidos mais adiantam é a semana 32(começo de Agosto).
'''
```

## Exercício 09
Investigue a relação entre o valor do frete e o preço dos produtos. Calcule a correlação entre esses dois fatores e crie um scatter plot para visualizar essa relação, discutindo como isso pode afetar as decisões de frete grátis.

```python
#fazendo uma correlação de pearson entre o preço do produto e o preço do frete
correlacao = df[['price', 'freight_value']].corr(method='pearson')
#fazendo um heatmap para visualizar a correlação
plt.figure( figsize=(7, 3) )
sns.heatmap( correlacao, annot=True, cbar=False )
plt.title('Correlação de Pearson entre o Preço do Produto e o Preço do Frete', fontdict={'fontsize': 11})

#fazendo uma correlação de spearman entre o preço do produto e o preço do frete
correlacao = df[['price', 'freight_value']].corr(method='spearman')
#fazendo um heatmap para visualizar a correlação
plt.figure( figsize=(7, 3) )
sns.heatmap( correlacao, annot=True, cbar=False )
plt.title('Correlação de Spearman entre o Preço do Produto e o Preço do Frete', fontdict={'fontsize': 11})

#criando uma nova coluna para armazenar a proporção entre o preço do frete e o preço do produto
df['proporcaoFrete/Preco'] = df['freight_value'] / df['price']

#fazendo uma correlação de pearson entre a proporção entre o preço do frete e do produto e o preço do produto
correlacao = df[['proporcaoFrete/Preco', 'price']].corr(method='pearson')
#fazendo um heatmap para visualizar a correlação
plt.figure( figsize=(7, 3) )
sns.heatmap( correlacao, annot=True, cbar=False )
plt.title('Correlação de Pearson entre Frete/Preço e o Preço do Produto', fontdict={'fontsize': 11})

#fazendo uma correlação de spearman entre a proporção entre o preço do frete e do produto e o preço do produto
correlacao = df[['proporcaoFrete/Preco', 'price']].corr(method='spearman')
#fazendo um heatmap para visualizar a correlação
plt.figure( figsize=(7, 3) )
sns.heatmap( correlacao, annot=True, cbar=False )
plt.title('Correlação de Spearman entre Frete/Preço e o Preço do Produto', fontdict={'fontsize': 11})

#fazendo um scatterplot para visualizar a relação entre o preço do produto e o preço do frete
plt.figure( figsize=(12, 6) )
sns.scatterplot( data=df, x='price', y='freight_value')
plt.title('Relação entre o Preço do Produto e o Preço do Frete')
plt.xlabel('Preço do Produto')
plt.ylabel('Preço do Frete')
plt.grid(True)

#fazendo um scatterplot para visualizar a relação entre a proporção entre o preço do frete e do produto e o preço do produto
plt.figure( figsize=(12, 6) )
sns.scatterplot( data=df, x='price', y='proporcaoFrete/Preco')
plt.title('Relação entre a Proporção entre o Preço do Frete e do Produto e o Preço do Produto')
plt.xlabel('Preço do Produto')
plt.ylabel('Proporção Frete/Preço')
plt.grid(True)


'''
Pela análise da correlação de Pearson e de Spearman entre o preço do produto e o preço do frete diretamente, podemos
observar que há uma correlação moderada entre essas duas variáveis, ou seja, à medida que uma variável aumenta, a outra
tende a aumentar, mas não de forma tão previsível. 

Por outro lado, ao analisar a correlação de Spearman entre a proporção entre o preço do frete e o preço do produto e o
preço do produto, podemos observar que a correlação é muito forte, ou seja, à medida que o preço do produto aumenta, a
proporção entre o preço do frete e o preço do produto tende a diminuir (de forma não linear, por isso utilizamos a
correlação de Spearman), o que para a nossa análise de auxiliar as decisões de frete grátis é muito relevante, já que,
para produtos mais caros, a empresa arcará com um pequeno custo de frete (em comparação ao preço do produto), mas que
pode atrair mais clientes e aumentar o volume de vendas, pois, muitas vezes, o frete grátis é um fator bem atraente
para os consumidores.
'''
```

Análises de feedback de clientes são essenciais para melhorias. Agrupe os dados de pedidos por avaliação média e calcule o volume de vendas para cada grupo, usando um gráfico de linha para mostrar a tendência de vendas em relação às avaliações.

```python
#carregando o dataset de avaliações
reviews_df = pd.read_csv('olist_order_reviews_dataset.csv')

#combinando o dataset de avaliações com o dataset principal
df = pd.merge(df, reviews_df, on='order_id')

#verificando a quantidade de valores duplicados
print(df.duplicated().sum())

print('\n------------------------------------\n')

#verificando e analisando a quantidade de valores nulos
print(df.isnull().sum())

'''
Em relação aos valores nulos nas colunas de avaliação, acredito que são normais, uma vez que os nulos aparecem
apenas em colunas em que são aceitáveis, que são as de comentário do cliente e título da avaliação.
'''

#calculando a média das avaliações por pedido
df_agrupado = df.groupby('order_id')['review_score'].mean()

#calculando o volume de vendas para cada grupo de avaliação
df_agrupado = df_agrupado.value_counts().sort_index()

#criando o gráfico de linhas
plt.figure(figsize=(12, 6))
plt.plot(df_agrupado.index, df_agrupado.values, marker='o')

#formatando o gráfico
plt.title('Quantidade de pedidos por avaliação')
plt.xlabel('Nota média da avaliação')
plt.ylabel('Quantidade de pedidos')
plt.grid(True)

#exibindo o gráfico
plt.show()
'''
Pelo gráfico é possível perceber que há uma tendência de haver mais vendas quando há avaliações com notas mais altas.
'''
```
